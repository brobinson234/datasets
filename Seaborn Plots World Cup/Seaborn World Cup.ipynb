```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('WorldCupMatches.csv')
#df_goals = pd.read_csv('goals.csv') # This dataset was no longer available when creating this Notebook. The dataset was only used in the final graph and does not affect the first two. Instead, I recreated the information by creating a new column within the WorldCupMatches DataFrame (df) and edited the code.

#print(df.head())

df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']

print(df.head())
total_goals = df.groupby('Year')['Total Goals'].sum().reset_index()

#print(total_goals)
sns.set_style('white')
sns.set_context('poster', font_scale=.22)
ax = plt.subplot()
ax = sns.barplot(data=total_goals, x='Year', y='Total Goals')
ax.set_title('Total Goals Per World Cup')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.set_context('poster', font_scale=.2)
ax2 = plt.subplot()
ax2 = sns.barplot(x=df['Year'], y=df['Total Goals'])
ax2.set_title('Total Goals per Game Over Time')
plt.show() # This graph automatically uses Seaborns aggregate feature to give the avg goals per game for the given year. This is not the same info as in the first graph, but it does cut out some of the intermediate steps that would be necessary if I was using Matplotlib to render this graph.

plt.clf()
```

```python
# Now to work on the third graph that will draw data from the newly created column to recreate the missing csv file. 
sns.set_style('whitegrid')
sns.set_context('notebook', font_scale=.4)
ax3 = plt.subplot()
ax3 = sns.boxplot(data=df, x='Year', y='Total Goals', palette='Spectral')
ax3.set_title('Boxplot Visualization of Goals per game over time')
plt.show()
```
