```python
import pandas as pd
import numpy as np 

from matplotlib import pyplot as plt 

from sklearn.linear_model import LinearRegression
```

```python
honey_data = pd.read_csv('honeyproduction.csv')
```

```python
print(honey_data.head())
```

```python
honey_data_mean = honey_data.groupby('year').totalprod.mean().reset_index()
print(honey_data_mean)
```

```python
X = honey_data_mean['year']
print(type(X))
print(X)
X = X.values.reshape(-1,1)
print(type(X))
print(X)
```

```python
y = honey_data_mean['totalprod']
print(y)
y = y.values.reshape(-1,1)
print(y)
```

```python
reg = LinearRegression()
reg = reg.fit(X,y)
intercept = reg.intercept_
print(intercept)
reg_coef = reg.coef_
print(reg_coef)
y_predict = reg.predict(X)
```

```python
plt.subplot()
plt.scatter(X,y)
plt.plot(X, y_predict)
plt.show()
```

```python
X_future = np.array(range(2013,2051))
print(X_future)
X_future = X_future.reshape(-1,1)
print(X_future)
```

```python
future_predict = reg.predict(X_future)
print(future_predict)
```

```python
plt.subplot()
plt.scatter(X,y)
plt.plot(X,y_predict)
plt.plot(X_future,future_predict)
plt.show()
```
